```python
# Intro to Seaborn visualtion package
# Lives on top of MatPlotLib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
# Plotting in line with the code as opposed to a separate window
plt.rcParams['figure.figsize'] = 8,4 # Plot size 8x4 to better utilize space on screen
```

```python
# Warnings are not important for us. 
# They don't really crash anything. Just more of heads-up. So we clean em up.
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
stats=pd.read_csv('P4-Demographic-Data.csv')
stats.columns=['CountryName','CountryCode', 'BirthRate', 'InternetUsers', 'IncomeGroup']
stats.head()
```

```python
# Distribution
vis1=sns.distplot(stats.Internetusers)
```

```python
# Distribution with 30 bins
vis1=sns.distplot(stats.Internetusers, bins=30)
```

```python
# Boxplot with a default value for data and defined axis names
vis2=sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
```

```python
# "Seaborn gallery" is great for finding example plots
```

```python
# Linear model plot without a linear regression line
vis3 = sns.lmplot(data = stats, x= 'InternetUsers', y='BirthRate', fit_reg=False)
```

```python
# Linear model plot without a linear regression line and color value from column 'IncomeGroup'
vis4 = sns.lmplot(data = stats, x= 'InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup')
```

```python
# Linear model plot 
# without a linear regression line and 
# color value from column 'IncomeGroup'
# size smaller and wider
vis5 = sns.lmplot(data = stats, x= 'InternetUsers', y='BirthRate', \
    fit_reg=False, hue='IncomeGroup', size=6, aspect =2)
```

```python
# Keyword arguments
# Goal is to change markers but we do it the difficult way
```

```python
# Marker size
# Seaborn does not specify a parameter for datapoint size
# Seaborn lmplot is a wrapper around matplotlib.pyplot.scatter()
# Seaborn specifies that scatter_kws (type: dict) is passed on to matplotlib.pyplot.scatter()
# From matplotlib docs we find that s is a parameter for datapoint size
# We pass a library entry 's':100 for the matplotlib.pyplot.scatter() parameter 's'
vis5 = sns.lmplot(data = stats, x= 'InternetUsers', y='BirthRate', \
    fit_reg=False, hue='IncomeGroup', size=6, aspect =2, \
    scatter_kws={'s':100})
```
